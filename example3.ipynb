```python
from seq2sec.apply import Protein
from seq2sec.model import load
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib notebook
```

```python
# load model and set to eval
net = load('./models/resnet-uncloss-dropblock-simpleblock-ss3_ss4_buried.pth') 
# net = load('./models/resnet_mt_3-4_cath95.pth')
```

```python
# read fasta
p = Protein("./fasta_sequences/start2fold/P0A7Y4.fasta")
# tmp = list(p.seq)
# tmp[4] = 'R'
# p.seq = ''.join(tmp)
p.sample2_with(net, n=100)
```

```python
plt.subplots(figsize=(len(p.seq)*0.20, 3))
plt.plot(np.transpose(p.buried));
plt.xticks(range(len(p.seq)),list(p.seq));
plt.yticks(np.arange(0,301,25));
```

```python
mean = np.mean(p.buried, axis=0)
std = np.std(p.buried, axis=0)
var = np.var(p.buried, axis=0)
plt.subplots(figsize=(len(p.seq)*0.20, 3))
plt.plot(mean)
plt.plot(mean-std)
plt.plot(mean+std)
plt.xticks(range(len(p.seq)),list(p.seq));
plt.yticks(np.arange(0,301,25));
```
